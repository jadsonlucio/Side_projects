```python
# iniciar dicionario

def split_space(text):
    cont_char=0
    key=""
    result=""
    text=text.replace("\t"," ").replace("\n","")
    while(cont_char<len(text)):
        if(text[cont_char]==" "):
            key=text[:cont_char]
            result=text[cont_char:]
            break
        
        cont_char=cont_char+1
    return key,result
    
dict_teste={}

for line in open("files//dicionario dataset.txt"):
    key,descricao=split_space(line)
    dict_teste[key]=descricao
    

```
